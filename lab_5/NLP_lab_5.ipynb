```python
from transformers import pipeline

unmasker_bert = pipeline('fill-mask', model='bert-base-multilingual-cased')
unmasker_xlm_roberta = pipeline('fill-mask', model='xlm-roberta-base')
unmasker_mt5 = pipeline('fill-mask', model='facebook/mbart-large-50')
```

**Task 3**

Devise a method to test if the langage model understands Polish cases. E.g. testing for nominal case could be expressed as "Warszawa to największe [MASK]", and the masked word should be in nominative case. Create sentences for each case.

**bert-base-multilingual-cased model**

```python
from pprint import pprint

def get_sentences(mask: str, task_number: int) -> list:

  if task_number == 3:
    sentences = [
        f'Warszawa to największe {mask} w Polsce.',
        f'Czy pojechałbyś kiedykolwiek do tego {mask}?',
        f'Przyglądam się temu {mask}.',
        f'Czuję wielką więź z tym {mask}.',
        f'Nie mogę przestać mówić o tym {mask}.',
        f'Ależ jest to piękne {mask}!',
        f'Idąc parkiem, ujrzałem piękne {mask}.',
        f'W parku nie było żadnych pięknych {mask}.',
        f'Przyglądam się tym pięknym {mask}.',
        f'Idę z {mask} na spacer.',
        f'Nie mogę przestać mówić o tych wspaniałych {mask}.',
        f'Ale zjawiskowe {mask}!'
    ]
  elif task_number == 4:
    sentences = [
        f'Wczoraj Kamil poszedł na siłownię i {mask} bicepsy.',
        f'Karolina usmażyła jajecznicę, a następnie ją {mask}.',
        f'Agata była głodna, więc {mask} jajecznicę.',
        f'Kacper otworzył drzwi Barbarze w restauracji, za co ona {mask} za obiad.',
        f'Poszedłem na spacer i {mask} piękną dziewczynę.'
    ]
  elif task_number == 5:
    sentences = [
        f'{mask} wrze w temperaturze 100 stopni, a zamarza w temperaturze 0 stopni Celsjusza.',
        f'Stolicą Polski jest {mask}.',
        f'{mask} to najwyższe góry w Polsce.',
        f'{mask} to najwyższe góry na świecie.',
        f'Wynikiem pomnożenia 2 przez 3 jest {mask}.',
        f'Liczby mające jako dzielnik tylko siebie i jeden to liczby {mask}.'
    ]


  return sentences
```

```python
for sentence in get_sentences('[MASK]', task_number=3):
  print(f'\nSentence: {sentence}\n')
  for idx in range(5):
    pprint(
        unmasker_bert(
            sentence
        )[idx]['sequence']
    )
```

**xlm-roberta-base model**


```python
for sentence in get_sentences('<mask>', task_number=3):
  print(f'\nSentence: {sentence}\n')
  for idx in range(5):
    pprint(
        unmasker_xlm_roberta(
            sentence
        )[idx]['sequence']
    )
```

**facebook/mbart-large-50 model**

```python
for sentence in get_sentences('<mask>', task_number=3):
  print(f'\nSentence: {sentence}\n')
  for idx in range(5):
    pprint(
        unmasker_mt5(
            sentence
        )[idx]['sequence']
    )
```
